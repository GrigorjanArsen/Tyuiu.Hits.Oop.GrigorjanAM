<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Product в C#, который будет представлять информацию о
продуктах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.
Требования к базовому классу Product:
• Атрибуты: Название (Name), Цена (Price), Производитель (Manufacturer).
• Методы:
o GetInfo(): метод для получения информации о продукте в виде строки.
o Discount(): метод для применения скидки к цене продукта.
o Display(): метод для отображения информации о продукте на экране.
Требования к производным классам:
1. Электроника (Electronics): Должен содержать дополнительные атрибуты,
такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть
переопределен для добавления логики учета гарантийного срока при
применении скидки.
2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как
Размер (Size). Метод Display() должен быть переопределен для добавления
информации о размере при отображении информации о продукте.
3. Книги (Books) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен
быть переопределен для включения информации об авторе в описании
продукта.

```
public interface IProduct
{
    string Availability(string avail);
}

public abstract class Product //новое
{
    public string Name { get; set; }
    public double Price { get; set; }
    public string Manufacturer { get; set; }
    public bool buyFlag{get;set;}
    public void CheckPrice(double price)
    {
        price = Price;
        Console.WriteLine($"Цена товара: {price}");
    }

    
    public void Buy(bool isbuy) //перегрузка 
    {
        if (isbuy)
        {
            Console.WriteLine("Вы собираетесь купить товар.");
            Console.WriteLine("Укажите количество товаров для покупки: ");
            buyFlag = true;
        }
        else
        {
            Console.WriteLine("Отказ от покупки");
        }
    }

    public void Buy(int count) //перегрузка 
    {
        if (buyFlag && count>0)
        {
            Console.WriteLine("'"+ count+ "'");
            Console.WriteLine($"Товары добавлены в корзину в количестве: {count}");
        }
        else if (!buyFlag)
        {
            Console.WriteLine("Вы не можете указать количество товаров, т.к. отказались от покупки");
        }
        else if(buyFlag && count<=0)
        {
            Console.WriteLine("Вы не можете купить 0 товаров или менее");
        }
    }

    public virtual int Availability(int num)
    {
        return num;
    }
    //Переопределить оператор + : когда к экземпляру прибавляется другой экземпляр, должна возвращаться новая сводка, объединяя списки обоих товаров
    public static int TotalProducts {get; private set;} = 0;
    public static double TotalPrice {get; private set;} = 0; //новое
    private bool IsPaid {get;set;} = false; //новое

    public Product(string name, double price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        TotalProducts++;
        TotalPrice += Price; //новое
    }
    public virtual string GetInfo()
    {
        string paymentStatus = IsPaid ? "Оплачено" : "Не оплачено"; //новое
        if (IsPaid)
        {
            DateTime purchaseDate = DateTime.Now; //новое
            return $"Название: {Name}, Цена: {Price}, Статус: {(IsPaid? "Оплачено" : "Не оплачено")}, Дата покупки: {purchaseDate}, Производитель: {Manufacturer}";
        }
        else
        {
            return $"Название: {Name}, Цена: {Price}, Статус: {(IsPaid? "Оплачено" : "Не оплачено")}, Производитель: {Manufacturer}";
        }
        
    }

    public virtual void Discount(double DiscPerc)
    {
        Price -= Price * (DiscPerc / 100);
    }

    public void Pay() //новое
    {
        IsPaid = true;
    }

    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    public static void DisplayTotalProducts()
    {
        Console.WriteLine($"Общее количество товаров: {TotalProducts}");
    }

    public static void DisplayTotalPrice() //новое
    {
        Console.WriteLine($"Общая стоимость товаров (без учета скидок): {TotalPrice}");
    }
}



public interface IFurniture //новое
{
    public int YearOfCreate{get; set;}
    void DisplayTotalFur();
}

public class AltClass : Product, IProduct
{
    public AltClass(string name, double price, string manufacturer) : base(name, price, manufacturer)
    {

    }

    public string Availability(string isaval)
    {
        return isaval;
    }

    public override int Availability(int avala)
    {
        return avala;
    }
}

public class Furniture : Product, IFurniture // множественное наследование
{
    public int YearOfCreate{get;set;} //новое
    private static int TotalFur { get; set; } = 0; //новое
    
    public Furniture(string name, double price, string manufacturer, int yearoc ) : base(name, price, manufacturer)
    {
        YearOfCreate = yearoc;
        TotalFur++;
    }
    public void DisplayTotalFur() //новое
    {
        Console.WriteLine($"Общее количество мебели: {TotalFur}");
    }
    public override string GetInfo() //новое
    {
        return base.GetInfo() + $", Год создания: {YearOfCreate}";
    }
}

public class Electronics : Product
{

    public static int TotalElectronics {get; private set;} = 0; //новое

    public Electronics(string name, double price, string manufacturer, int warrantyPeriod) : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
        TotalElectronics++;
    }
    public int WarrantyPeriod { get; set; }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Гарантийный срок: {WarrantyPeriod} месяцев";
    }

    public override void Discount(double DiscPerc)
    {
        if (WarrantyPeriod < 3)
        {
            DiscPerc *= 1.2;
        }
        base.Discount(DiscPerc);
    }

    public static void DisplayTE() //новое
    {
        Console.WriteLine($"Общее количество электроники: {TotalElectronics}");
    }

    public override void Display()
    {
        Console.WriteLine(GetInfo());
    }

    
}

public class Clothing : Product
{
    public int Size { get; set; }
    public string Material {get; set;} //новое
    public string Cut {get;set;} //новое
    public static int TotalClothing {get; private set;} = 0; //новое

    public Clothing(string name, double price, string manufacturer, int size, string material, string cut) : base(name, price, manufacturer)
    {
        Size = size;
        Material = material;
        Cut = cut;
        TotalClothing++;
    }

    public override void Discount(double DiscPerc)  //новое
    {
        DiscPerc = DiscPerc + TotalClothing;
        base.Discount((double)DiscPerc);
    }

    public static void DisplayTotalCloth()  //новое
    {
        Console.WriteLine($"Количество одежды : {TotalClothing}");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Размер: {Size}" + $", Материал: {Material}" + $", Крой: {Cut}";
    }

    public override void Display()
    {
        Console.WriteLine(GetInfo());
    }
}

public class Books : Product
{
    public string Author { get; set; }
    public static int TotalBooks {get; private set;} = 0; //новое
    protected int PagesQuantity {get; set;} //новое
    public Books(string name, double price, string manufacturer, string author, int pagesquan) : base(name, price, manufacturer)
    {
        Author = author;
        TotalBooks++;
        PagesQuantity = pagesquan; //новое
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Автор: {Author}" + $", Количество страниц: {PagesQuantity}";
    }

    public override void Display()
    {
        Console.WriteLine(GetInfo());
    }
    public void DisplayTB() //новое
    {
        Console.WriteLine($"Общее количество книг: {TotalBooks}");
    }
}


public class AudioBooks : Books // сложное наследование product-books-audiobooks
{
    public new string Author { get; set; }
    public int Duration {get; set;} // новое
    public static int TotalAudioBooks{get; private set;} = 0; //новое
 
    public AudioBooks(string name, double price, string manufacturer, string author, int duration, int pagesquan) : base(name, price, manufacturer, author, pagesquan)
    {
        Author = author;
        Duration = duration;

        TotalAudioBooks++; //новое
    }
    public override string GetInfo()
    {
        return base.GetInfo() + $", Автор: {Author}" + $", Длительность 'мин.' : {Duration}";
    }

    public override void Display()
    {
        Console.WriteLine(GetInfo());
    }
    public void DisplayTAB() //новое
    {
        Console.WriteLine($"Общее количество аудиокниг: {TotalAudioBooks}");
    }
}

public interface IFood //новое
{
    public DateTime ExpirationDate { get; set; }
    bool IsFresh();
}
public class Food : Product, IFood  //новое множественное наследование от разных базовых классов
{
    public DateTime ExpirationDate { get; set; }
    public static int TotalFood {get; private set;} = 0; //новое

    public Food(string name, double price, string manufacturer, DateTime expirationDate) : base(name, price, manufacturer)
    {
        ExpirationDate = expirationDate;
        TotalFood++;
    }

    public bool IsFresh()
    {
        return DateTime.Now < ExpirationDate;
    }

    public override string GetInfo()
    {
        string freshnessInfo = IsFresh() ? "Продукт свежий" : "Срок годности продукта истёк";
        return base.GetInfo() + $", Срок годности: {ExpirationDate.ToShortDateString()} ({freshnessInfo})";
    }

    public void DisplayTotalFood() //новое
    {
        Console.WriteLine($"Общее количество продуктов питания: {TotalFood}");
    }
    public override void Display()
    {
        Console.WriteLine(GetInfo());
    }
}


public class ProductCollection<P> where P: Product
{

    public List<P> _products = new List<P>();

    public static ProductCollection<P> operator +(ProductCollection<P> productor1, ProductCollection<P> productor2)
    {
        ProductCollection<P> resultProductor = new ProductCollection<P>();
        resultProductor._products.AddRange(productor1._products);
        resultProductor._products.AddRange(productor2._products);

        foreach(var productor in resultProductor._products)
        {
            productor.Display();
        }
        return resultProductor;
    }

    public void Add(P product)
    {
        _products.Add(product);
    }
    
    public void Remove(P product)
    {
        _products.Remove(product);
    }

    public void DisplayProduct()
    {
        foreach(var product in _products)
        {
            product.Display();
        }
    }
    // private List<P> _resproductors = new List<P>();

    // public void toResultProductor(List<P> productor1, List<P> productor2)
    // {
    //     _resproductors.AddRange(productor1);
    //     _resproductors.AddRange(productor2);
        
    //     foreach (var productor in _resproductors)
    //     {
    //         productor.Display();
    //     }
    // }

    
}

ProductCollection<Product> productor1 = new ProductCollection<Product>();

productor1.Add(new Furniture("Стол деревянный", 2250, "Россия", 2023));
productor1.Add(new Clothing("Футболка белая однотонная", 750, "Бангладеш", 52, "Хлопок", "Оверсайз"));

ProductCollection<Product> productor2 = new ProductCollection<Product>();

productor2.Add(new Electronics("Телевизор LG", 30000, "Китай", 4));
productor2.Add(new Books("азбука", 12, "Россия", "Я",123));


ProductCollection<Product> resultProductor = productor1 + productor2;

// ProductCollection<Product> productor = new ProductCollection<Product>();
// productor.Add(new Furniture("Стол деревянный", 2250, "Россия", 2023));
// productor.Add(new Clothing("Футболка белая однотонная", 750, "Бангладеш", 52, "Хлопок", "Оверсайз"));
// productor.DisplayProduct();

// ProductCollection<Product> productor2 = new ProductCollection<Product>();
// productor2.Add(new Electronics("Телевизор LG", 30000, "Китай", 4));
// productor2.Add(new Books("азбука", 12, "Россия", "Я",123));


// Furniture Table = new Furniture("Стол деревянный", 2250, "Россия", 2023);
// Table.Discount(5);
// Table.Pay();
// Table.Display();
// Table.Buy(true);
// Table.Buy(0);

// Furniture Chair = new Furniture("Кресло кожаное", 3500, "Италия", 2022);
// Chair.Discount(10); 
// Chair.Pay();     
// Chair.Display();    
// Table.DisplayTotalFur();
// Console.WriteLine();

// Clothing Tee = new Clothing("Футболка белая однотонная", 750, "Бангладеш", 52, "Хлопок", "Оверсайз");
// Tee.Discount(12);
// Tee.Pay();
// Tee.Display();

// Clothing Hat = new Clothing("Бейсболка DC", 2000, "Германия", 60, "Вельвет", "Стандарт");
// Hat.Discount(1);
// Hat.Pay();
// Hat.Display();

// Clothing TShirt = new Clothing("Футболка DC", 800, "Германия", 40, "Полиэстер", "Компакт");
// TShirt.Discount(1);
// TShirt.Display();

// Clothing Coat = new Clothing("Пальто кожаное", 5000, "Италия", 48, "Кожа", "Стильный");
// Coat.Discount(1);
// Coat.Display();

// Clothing Shorts = new Clothing("Шорты спортивные", 300, "Китай", 32, "Полиэстер", "Атлетические");
// Shorts.Discount(1);
// Shorts.Pay();
// Shorts.Display();

// Clothing.DisplayTotalCloth();
// Console.WriteLine();

// Electronics tv = new Electronics("Телевизор LG", 30000, "Китай", 4);
// tv.Discount(4);
// tv.Pay();
// tv.Display();


// Electronics phone = new Electronics("Смартфон Vivo", 14000, "Китай", 1);
// phone.Discount(25);
// phone.Pay();
// phone.Display();

// Electronics.DisplayTE();
// Console.WriteLine();

// Books atlas = new Books("Атлас", 124, "Россия", "Афанасий Никитин", 241);
// atlas.Display();
// atlas.DisplayTB();
// Console.WriteLine();

// AudioBooks fantasy = new AudioBooks("Аудиокнига '20 тысяч лье под водой'", 172, "Россия", "Жюль Верн", 423,124);
// fantasy.Display();
// fantasy.DisplayTAB();

// Console.WriteLine();

// Food milk = new Food("Молоко", 90, "Беларусь", new DateTime(2024, 10, 30));
// milk.Discount(10);
// milk.Display();


// Food bread = new Food("Хлеб", 40, "Россия", new DateTime(2023, 9, 25));
// bread.Discount(5); 
// bread.Display();
// bread.DisplayTotalFood();
// Console.WriteLine();

Product.DisplayTotalProducts();
Product.DisplayTotalPrice();
```
